```python
import pyart
import matplotlib.pyplot as plt
from metpy.plots import ctables
```

```python
# Load NEXRAD Level 3 data file
filename = '/scratch2/BMC/fv3lam/Robby.Frost/radar_obs/20190520/kfdr/KOUN_SDUS54_N0QFDR_201905201601'
kfdr = pyart.io.read_nexrad_level3(filename)

# Plotting parameters
display = pyart.graph.RadarMapDisplay(kfdr)
# mapping
plt_area = [-104, -94, 30, 39] # W, E, S, N
# Set up the plot figure and map
fig = plt.figure(figsize=(10, 8))
# define color table for reflectivity
colors = ctables.registry.get_colortable('NWSReflectivity')
# Plot NEXRAD data
display.plot_ppi_map('reflectivity', 0, vmin=0, vmax=75.1,
                     resolution='10m', cmap=colors,
                     min_lat=30,max_lat=39,
                     min_lon=-104, max_lon=-94,
                     add_grid_lines=False)
# display radar location
display.plot_point(kfdr.longitude['data'][0], kfdr.latitude['data'][0], label_text='KFDR', symbol='ro')
# Show the plot
plt.show()
plt.close()
```
